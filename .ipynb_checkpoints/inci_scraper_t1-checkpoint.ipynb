```python
def fetch_inci_data(ingredient_name):
    url_name = ingredient_name.lower().replace(" ", "-")
    url = f"https://incidecoder.com/ingredients/{url_name}"
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"[{ingredient_name}] Not found.")
        return None

    soup = BeautifulSoup(response.text, "html.parser")

    data = {"name": ingredient_name}

    desc_tag = soup.select_one("div.molecule-description > p")
    if desc_tag:
        data["description"] = desc_tag.text.strip()

    synonym_tags = soup.select("div.synonyms span.tag")
    data["synonyms"] = [tag.text.strip() for tag in synonym_tags] if synonym_tags else []

    print(f"[{ingredient_name}] Scraped successfully.")
    return data

```

```python
import requests
from bs4 import BeautifulSoup
import json
import time

def scrape_ingredient(slug):
    url = f"https://incidecoder.com/ingredients/{slug}"
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    try:
        res = requests.get(url, headers=headers)
        if res.status_code != 200:
            return None

        soup = BeautifulSoup(res.text, "html.parser")
        name = soup.select_one("h1.ingredient-title").text.strip()
        desc_tag = soup.select_one("div.ingredient-description-text")
        description = desc_tag.text.strip() if desc_tag else "No description found"

        return {
            "name": name,
            "slug": slug,
            "description": description,
            "synonyms": [],     # Placeholder for now
            "risk": "Unknown"   # Will be enriched in Step 1.3
        }
    except Exception as e:
        print(f"[{slug}] Error: {e}")
        return None
```

```python
def scrape_ingredient(slug):
    url = f"https://incidecoder.com/ingredients/{slug}"
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    try:
        res = requests.get(url, headers=headers)
        if res.status_code != 200:
            print(f"[{slug}] ❌ HTTP {res.status_code}")
            return None

        soup = BeautifulSoup(res.text, "html.parser")
        title_tag = soup.select_one("h1.ingredient-title")
        if not title_tag:
            print(f"[{slug}] ❌ title not found, dumping snippet:")
            print(soup.prettify()[:1000])  # Just show 1000 chars for debug
            return None

        name = title_tag.text.strip()
        desc_tag = soup.select_one("div.ingredient-description-text")
        description = desc_tag.text.strip() if desc_tag else "No description found"

        return {
            "name": name,
            "slug": slug,
            "description": description,
            "synonyms": [],
            "risk": "Unknown"
        }
    except Exception as e:
        print(f"[{slug}] ❌ Error: {e}")
        return None
```

```python
# ✅ Step 1.2 ingredient list
ingredient_list = [
    "glycerin", "dimethicone", "tocopheryl-acetate", "aqua", "butylene-glycol", "ethylhexylglycerin",
    "niacinamide", "hyaluronic-acid", "allantoin", "panthenol", "ascorbic-acid", "zinc-oxide",
    "titanium-dioxide", "retinol", "salicylic-acid", "lactic-acid", "glycolic-acid", "citric-acid",
    "ceramide-np", "ceramide-ap", "ceramide-eop", "cetearyl-alcohol", "stearic-acid", "sodium-hyaluronate",
    "isopropyl-myristate", "phenoxyethanol", "fragrance", "alcohol", "benzyl-alcohol", "sodium-lauryl-sulfate",
    "sodium-laureth-sulfate", "disodium-edta", "squalane", "jojoba-oil", "avobenzone", "octinoxate",
    "octocrylene", "peg-100-stearate", "cetyl-alcohol", "carbomer", "triethanolamine", "retinyl-palmitate",
    "kaolin", "bentonite", "azelaic-acid", "tea-tree-oil", "licorice-root-extract", "green-tea-extract", "urea"
]

# Run scraper
results = []

for name in ingredient_list:
    data = scrape_ingredient(name)
    if data:
        results.append(data)
        print(f"[{name}] ✅ Scraped")
    else:
        print(f"[{name}] ❌ Failed")
    time.sleep(1.5)  # Add delay to avoid hammering the site

# Save output
with open("inci_enriched.json", "w") as f:
    json.dump(results, f, indent=2)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

def scrape_incidecoder_with_brave(slug):
    url = f"https://incidecoder.com/ingredients/{slug}"

    brave_path = "/Applications/Brave Browser.app/Contents/MacOS/Brave Browser"  # <-- UPDATE THIS for your OS

    options = Options()
    options.binary_location = brave_path
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    try:
        driver.get(url)
        time.sleep(3)  # Wait for JS to load

        soup = BeautifulSoup(driver.page_source, "html.parser")
        title = soup.select_one("h1.ingredient-title")
        desc = soup.select_one("div.ingredient-description-text")

        return {
            "name": title.text.strip() if title else "Not found",
            "description": desc.text.strip() if desc else "No description found"
        }
    finally:
        driver.quit()

# Test example
print(scrape_incidecoder_with_brave("glycerin"))
```
