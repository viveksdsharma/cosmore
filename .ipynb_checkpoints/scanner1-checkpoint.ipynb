```python
# OCR + Fuzzy Match Sample Pipeline for Cosmetic Ingredient Detection

import pytesseract
pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'

from PIL import Image
import re
from rapidfuzz import fuzz, process
```

```python
# Sample ingredient database (simulate DB lookup)
ingredient_db = {
    "methylparaben": {"risk": "Caution", "description": "Preservative; may cause irritation"},
    "glycerin": {"risk": "Safe", "description": "Humectant that hydrates skin"},
    "fragrance": {"risk": "Harmful", "description": "Contains allergens; poorly regulated"},
    "rose extract": {"risk": "Safe", "description": "Natural extract with soothing properties"},
    "water": {"risk": "Safe", "description": "Common base ingredient (aqua)"}
}

```

```python

# Step 1: OCR function
def extract_text_from_image(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)
    return text
```

```python

# Step 2: Clean and parse ingredients
def clean_and_split_ingredients(raw_text):
    raw_text = raw_text.lower()
    raw_text = re.sub(r'[^a-zA-Z0-9,\-/\n ]+', '', raw_text)
    lines = raw_text.split("\n")
    ingredients = []
    for line in lines:
        if 'ingredient' in line:
            line = re.sub(r'ingredients?:', '', line)
            parts = re.split(r'[,/]', line)
            ingredients += [p.strip() for p in parts if p.strip()]
    return ingredients

```

```python
def match_ingredients(ingredient_list):
    results = []
    for item in ingredient_list:
        result = process.extractOne(item, ingredient_db.keys(), scorer=fuzz.token_sort_ratio)
        if result and len(result) >= 2:
            match, score = result[:2]
            if score > 80:
                data = ingredient_db[match]
                results.append({"input": item, "match": match, "risk": data['risk'], "desc": data['description']})
                continue
        # fallback if no good match
        results.append({"input": item, "match": None, "risk": "Unknown", "desc": "Ingredient not found in database."})
    return results


```

```python
# Example usage (replace with image path)
image_path = '/Users/viveksharma/Desktop/programming/scanner/label1.png'
text = extract_text_from_image(image_path)
ingredient_list = clean_and_split_ingredients(text)
final_results = match_ingredients(ingredient_list)
print(final_results)
```
