```python
import re
from rapidfuzz import process, fuzz
from PIL import Image
import pytesseract

pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'

```

```python
# STEP 1: Normalize helper
def normalize(text):
    return re.sub(r'[^a-z0-9 ]+', '', text.lower().strip())

# STEP 2: OCR extraction
def extract_text_from_image(image_path):
    image = Image.open(image_path)
    return pytesseract.image_to_string(image)

```

```python
# STEP 3: Clean text into ingredient list
def clean_and_split_ingredients(text):
    lines = re.split(r'[,\n]', text.lower())
    ingredients = []
    for line in lines:
        parts = re.split(r'\band\b|\/|,', line)
        for part in parts:
            cleaned = part.strip(" -[]()")
            if cleaned:
                ingredients.append(cleaned)
    return ingredients

# STEP 4: Corrections for common OCR errors
corrections = {
    "wlantoin": "allantoin",
    "ethythexylglycerin": "ethylhexylglycerin",
    "glyceryl stearal": "glyceryl stearate",
    "butyrospermum": "butyrospermum parkii butter",
    "fragaria ananassa": "fragaria ananassa fruit extract",
    "rubus chamae": "rubus chamaemorus seed oil",
    "morus seed oil": "morus alba seed oil",
    "aqua ": "aqua",
    "ethythexylglycerin": "ethylhexylglycerin",
    "peg 100 stearate": "peg100 stearate",
    "butyrospermum ter": "butyrospermum parkii",
    "rou it extract": "fruit extract",
    "wlanotoin": "allantoin",
    "glyceryl stearal": "glyceryl stearate",
}

```

```python
# STEP 5: Ingredient database (30+ examples)
ingredient_db_raw = {
    "aqua": {"risk": "Safe", "description": "Water, used as a solvent"},
    "glycerin": {"risk": "Safe", "description": "Humectant that hydrates skin"},
    "dimethicone": {"risk": "Low", "description": "Silicone used for smoothing"},
    "caprylic capric triglyceride": {"risk": "Safe", "description": "Emollient derived from coconut oil"},
    "isohexadecane": {"risk": "Low", "description": "Emollient that gives silky feel"},
    "cetyl alcohol": {"risk": "Low", "description": "Fatty alcohol, moisturizing"},
    "cetearyl alcohol": {"risk": "Low", "description": "Emulsifier and stabilizer"},
    "glyceryl stearate": {"risk": "Low", "description": "Emulsifier"},
    "peg100 stearate": {"risk": "Moderate", "description": "Synthetic emulsifier"},
    "butyrospermum parkii butter": {"risk": "Safe", "description": "Shea Butter"},
    "xylitylglucoside": {"risk": "Safe", "description": "Sugar-derived humectant"},
    "anhydroxylitol": {"risk": "Safe", "description": "Moisturizer from sugar"},
    "butylene glycol": {"risk": "Low", "description": "Solvent and moisturizer"},
    "polysorbate 20": {"risk": "Low", "description": "Emulsifier"},
    "polysorbate 80": {"risk": "Low", "description": "Emulsifier"},
    "palmitoyl tetrapeptide7": {"risk": "Safe", "description": "Anti-aging peptide"},
    "fragaria ananassa fruit extract": {"risk": "Safe", "description": "Strawberry extract"},
    "sodium hyaluronate": {"risk": "Safe", "description": "Hydrating molecule"},
    "rubus chamaemorus seed oil": {"risk": "Safe", "description": "Cloudberry seed oil"},
    "panthenol": {"risk": "Safe", "description": "Provitamin B5, soothing"},
    "tocopheryl acetate": {"risk": "Low", "description": "Vitamin E derivative"},
    "allantoin": {"risk": "Safe", "description": "Skin-soothing agent"},
    "disodium edta": {"risk": "Moderate", "description": "Chelating agent"},
    "phenoxyethanol": {"risk": "Moderate", "description": "Preservative"},
    "ethylhexylglycerin": {"risk": "Low", "description": "Preservative booster"},
    "cyclopentasiloxane": {"risk": "Low", "description": "Volatile silicone"},
    "peg40 hydrogenated castor oil": {"risk": "Moderate", "description": "Emulsifier"},
    "propylene glycol": {"risk": "Low", "description": "Solvent and moisturizer"},
    "niacinamide": {"risk": "Safe", "description": "Brightening and anti-inflammatory"},
    "triethanolamine": {"risk": "Moderate", "description": "pH adjuster"},
    "squalane": {"risk": "Safe", "description": "Emollient"},
    "betaine": {"risk": "Safe", "description": "Hydrating amino acid"},
}

ingredient_db = {normalize(k): v for k, v in ingredient_db_raw.items()}
```

```python

# STEP 6: Matching logic
def match_ingredients(ingredient_list):
    results = []
    for item in ingredient_list:
        norm_item = corrections.get(item, item)
        match = process.extractOne(norm_item, ingredient_db.keys(), scorer=fuzz.token_sort_ratio)
        if match:
            match_text, score = match[:2]
            if score > 78:
                data = ingredient_db[match_text]
                results.append({"input": item, "match": match_text, "risk": data['risk'], "desc": data['description']})
                continue
        results.append({"input": item, "match": None, "risk": "Unknown", "desc": "Ingredient not found in database."})
    return results

```

```python
# STEP 7: Run it
image_path = '/Users/viveksharma/Desktop/programming/scanner/label1.png'  # image path
text = extract_text_from_image(image_path)
ingredient_list = clean_and_split_ingredients(text)
final_results = match_ingredients(ingredient_list)

for r in final_results:
    print(r)
```
